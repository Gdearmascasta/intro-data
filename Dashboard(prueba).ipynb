```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact, Dropdown
import scienceplots
import warnings
plt.style.use(["science","notebook","grid"])
warnings.filterwarnings("ignore")


titanic = sns.load_dataset("titanic")
titanic.dropna(subset=['age'], inplace=True)


def age_group(age): # Crear columna de rango de edad
    if age <= 17:
        return "0-17"
    elif 18 <= age <= 35:
        return "18-35"
    elif 36 <= age <= 55:
        return "36-55"
    else:
        return "56+"

titanic["age_range"] = titanic["age"].apply(age_group)

def passenger_category(row): # Crear columna de categoría: niño, hombre, mujer
    if row['age'] <= 17:
        return "niño"
    else:
        return row['sex']  # hombre o mujer

titanic["category"] = titanic.apply(passenger_category, axis=1)


def dashboard(clase, edad): # Función para graficar

    if clase != "Todos":# Filtrar por clase
        data = titanic[titanic['class'] == clase]
    else:
        data = titanic.copy()

    if edad != "Todos":  # Filtrar por edad
        data = data[data['age_range'] == edad]



    titulo = f"Dashboard Titanic - Clase: {clase}, Edad: {edad}" # Título dinámico


    fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # Crear subplots
    fig.suptitle(titulo, fontsize=16, fontweight="bold")


    sns.violinplot(data=data, x="sex", y="fare", ax=axes[0], palette="Set2")  # Gráfico de violín: tarifas por sexo
    axes[0].set_title("Distribución de tarifas por sexo")

    sns.countplot(data=data, x="category", hue="survived", ax=axes[1], palette="Set1") # Countplot: supervivencia por categoría
    axes[1].set_title("Supervivencia por categoría (niño, hombre, mujer)")
    axes[1].legend(title="Sobrevivió", labels=["No", "Sí"])

    plt.show()

# Widgets Interactivos
clase_dropdown = Dropdown(
    options=["Todos", "First", "Second", "Third"],
    description="Clase:"
)

edad_dropdown = Dropdown(
    options=["Todos", "0-17", "18-35", "36-55", "56+"],
    value="Todos",
    description="Edad:"
)

interact(dashboard, clase=clase_dropdown, edad=edad_dropdown)
```

1) En la Primera Clase se observa que las mujeres presentan una dispersión mucho mayor en las tarifas pagadas que los hombres, ya que sus boletos abarcan desde precios bajos hasta los más elevados de todo el barco, lo cual indica que había tanto pasajeras en cabinas sencillas como en suites de lujo; en contraste, los hombres de 1ª Clase muestran una distribución más estrecha en los precios, lo que refleja menos variabilidad en el costo de sus pasajes.


2) Los niños de 1ª Clase tuvieron una probabilidad de supervivencia mucho mayor que los de 3ª Clase, pues la mayoría de ellos logró acceder a los botes salvavidas gracias a la prioridad de “mujeres y niños primero” y a su ubicación más cercana a las cubiertas de evacuación, mientras que en 3ª Clase la situación fue mucho más desfavorable: el acceso a los botes estaba restringido por la distancia, la confusión y las condiciones físicas del barco, lo que redujo drásticamente las posibilidades de salvar a los menores.

3) En la Segunda Clase se nota una diferencia entre los jóvenes adultos (18-35) y los adultos (36-55), ya que los primeros sobrevivieron en una proporción ligeramente mayor que los segundos, lo cual puede explicarse porque la edad más temprana les daba mejores condiciones físicas para moverse con rapidez en medio del caos y posiblemente más oportunidades de ser rescatados, mientras que los adultos de mayor edad mostraron una tasa de supervivencia aún más baja, evidenciando que este grupo etario tuvo menos probabilidades de salvarse.

4) La diferencia más drástica entre hombres y mujeres se presenta en la Tercera Clase dentro del rango de edad de 18 a 35 años, donde la mayoría de las mujeres jóvenes logró sobrevivir mientras que la gran mayoría de los hombres murió; esta marcada disparidad muestra cómo influyeron las normas sociales de la época, ya que incluso en condiciones difíciles se priorizó la evacuación de mujeres, dejando a los hombres jóvenes de las clases más bajas con una mínima posibilidad de salvarse.
