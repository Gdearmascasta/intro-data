```python
# Datos de ejemplo (puedes reemplazarlos por los reales)
years = ["2017", "2018", "2019", "2020", "2021", "Nov-21", "Dec-21", "Jan-22", "2022 Q1", "2022 Q2"]
loan_loss_reserves = [1.25, 1.3, 1.55, 1.6, 1.65, 1.7, 1.65, 1.5, 1.4, 1.42]
npls = [0.9, 0.8, 1.2, 1.25, 1.5, 1.65, 1.5, 1.6, 1.45, 1.42]
loan_loss_reserves_pct = [6.8, 6.8, 7.2, 7.0, 7.5, 7.5, 7.5, 7.2, 6.5, 6.5]
npl_rate = [4.0, 3.5, 4.5, 4.4, 5.2, 5.8, 5.8, 5.2, 5.5, 5.0]

x = np.arange(len(years))  # posiciones en el eje X
width = 0.35  # ancho de las barras

fig, ax1 = plt.subplots(figsize=(10,6))

# Eje primario (izquierda, $ billion)
bars1 = ax1.bar(x - width/2, loan_loss_reserves, width, label="Loan Loss Reserves", color="royalblue")
bars2 = ax1.bar(x + width/2, npls, width, label="NPLs", color="darkorange")
ax1.set_ylabel("$ billion")
ax1.set_xticks(x)
ax1.set_xticklabels(years, rotation=30)

# Eje secundario (derecha, %)
ax2 = ax1.twinx()
ax2.plot(x, loan_loss_reserves_pct, color="green", marker="o", label="Loan Loss Reserves %")
ax2.plot(x, npl_rate, color="gold", marker="o", label="NPL Rate")
ax2.set_ylabel("%")

# Título y leyenda
fig.suptitle("NPLs and Loan Loss Reserves")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

```python
# Get the latest year
latest_year = x1.Year.max()

# Get Colombia's rank in the latest year
colombia_latest_rank = x1[(x1["Country name"] == "Colombia") & (x1["Year"] == latest_year)]["Rank"].iloc[0]
colombia_latest_score = x1[(x1["Country name"] == "Colombia") & (x1["Year"] == latest_year)]["Ladder score"].iloc[0]

# Find countries with higher and lower rank (lower rank means higher happiness)
happier_countries = x1[(x1["Year"] == latest_year) & (x1["Rank"] < colombia_latest_rank)].sort_values(by="Rank").head(2)["Country name"].tolist()
less_happy_countries = x1[(x1["Year"] == latest_year) & (x1["Rank"] > colombia_latest_rank)].sort_values(by="Rank").tail(2)["Country name"].tolist()

# Combine Colombia with the selected countries
selected_countries = happier_countries + ["Colombia"] + less_happy_countries

print(f"Selected countries: {selected_countries}")
```

```python
filtered_df = x1[x1['Country name'].isin(selected_countries)]
display(filtered_df.head())
```

```python
plt.figure(figsize=(12, 7))

for country in selected_countries:
    country_data = filtered_df[filtered_df['Country name'] == country].sort_values(by='Year')
    plt.plot(country_data['Year'], country_data['Ladder score'], marker='o', label=country)

plt.xlabel('Year')
plt.ylabel('Ladder score')
plt.title('Happiness Score Over Time for Selected Countries')
plt.legend()
plt.grid(True)
plt.show()
```
